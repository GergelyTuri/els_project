```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
moseq = pd.read_csv('/Users/novak/Documents/Columbia/mentored_research/Turi/kpms_data_for_daart/ptsd9_recall4_28-4DLC_resnet50_phi_cfc_boxNov7shuffle1_388000.csv')
moseq.head()
```

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(moseq[['syllable']])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['syllable']))
```

```python
background_name_num = [8, 10, 11, 15, 16, 17, 18, 19, 20, 21, 23, 24, 25, 26, 27, 28, 29, 31, 32, 36, 37, 38, 39, 40, 46, 49, 52, 53, 54, 57, 58, 59, 60, 62, 71, 73, 76, 86, 91, 92]
drop_name = []
for i in background_name_num:
    drop_name.append('syllable_' + str(i))
drop_index = encoded_df.columns.get_indexer(drop_name)
drop_index = [x for x in drop_index if x != -1]
drop_df = encoded_df.iloc[:, drop_index]
```

```python
background_df = drop_df.max(axis=1)
keep_df = encoded_df.drop(columns=encoded_df.columns[drop_index], axis=1)
final_df = pd.concat([keep_df, background_df], axis=1)

final_df = final_df.rename(columns={0: 'background'})

final_df = final_df[sorted(final_df.columns)]

final_df
```
